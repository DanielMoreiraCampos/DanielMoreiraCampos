```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import seaborn as sns
```

```python
#stats 
df_inPerson = df[(df['timeperiod'] <= 2)]
df_inPerson.iloc[0:8400,[11,12,13,14,15,16]].describe()
```

```python
r_mean = df_inPerson['readingscore'].mean()
rSL_mean = df_inPerson['readingscoreSL'].mean()
w_mean = df_inPerson['writingscore'].mean()
wSL_mean = df_inPerson['writingscoreSL'].mean()
m_mean = df_inPerson['mathscore'].mean()
mSL_mean = df_inPerson['mathscoreSL'].mean()

exames = ['Reading', 'Reading State Level', 'Writing', 'Writing State Level', 'Math', 'Math State Level']
grades = [r_mean, rSL_mean, w_mean, wSL_mean, m_mean, mSL_mean]
colors = ['darkslateblue','mediumslateblue','darkslateblue','mediumslateblue','darkslateblue', 'mediumslateblue']

#plt.bar(dict_lang.keys(), dict_lang.values())
plt.figure(figsize=(10,6))
plt.barh(exames,grades, color=colors)
plt.title('Knowledge of Programming Languages')
```

```python
#stats 
df_online = df[(df['timeperiod'] > 2)]
df_online.iloc[0:8400,[11,12,13,14,15,16]].describe()
```

```python
r_mean = df_online['readingscore'].mean()
rSL_mean = df_online['readingscoreSL'].mean()
w_mean = df_online['writingscore'].mean()
wSL_mean = df_online['writingscoreSL'].mean()
m_mean = df_online['mathscore'].mean()
mSL_mean = df_online['mathscoreSL'].mean()

exames = ['Reading', 'Reading State Level', 'Writing', 'Writing State Level', 'Math', 'Math State Level']
grades = [r_mean, rSL_mean, w_mean, wSL_mean, m_mean, mSL_mean]
colors = ['mediumseagreen','mediumaquamarine','mediumseagreen','mediumaquamarine','mediumseagreen', 'mediumaquamarine']

#plt.bar(dict_lang.keys(), dict_lang.values())
plt.figure(figsize=(10,6))
plt.barh(exames,grades, color=colors)
plt.title('Knowledge of Programming Languages')
```

```python
F_r= df_inPerson[ (df_inPerson['readingscore'] >= 0) & (df_inPerson['readingscore'] <= 59)]
D_r= df_inPerson[ (df_inPerson['readingscore'] >= 60) & (df_inPerson['readingscore'] <= 69)]
C_r= df_inPerson[ (df_inPerson['readingscore'] >= 70) & (df_inPerson['readingscore'] <= 79)]
B_r= df_inPerson[ (df_inPerson['readingscore'] >= 80) & (df_inPerson['readingscore'] <= 89)]
A_r= df_inPerson[ (df_inPerson['readingscore'] >= 90) & (df_inPerson['readingscore'] <= 100)]

per_F = F_r['readingscore'].sum() / df_inPerson['readingscore'].sum()*100
per_D = D_r['readingscore'].sum() / df_inPerson['readingscore'].sum()*100
per_C = C_r['readingscore'].sum() / df_inPerson['readingscore'].sum()*100
per_B = B_r['readingscore'].sum() / df_inPerson['readingscore'].sum()*100
per_A = A_r['readingscore'].sum() / df_inPerson['readingscore'].sum()*100
grades=[per_F, per_D, per_C, per_B, per_A]

color = ['red', 'orange', 'yellow', 'blue', 'green']
explode = (0.05, 0.05, 0.05, 0.05, 0.05)
labels = ['0->59', '60->69', '70->79', '80->89', '90->100']
plt.pie(grades, colors=color, explode=explode, labels=labels, shadow=True, autopct='%1.1f%%')
```
